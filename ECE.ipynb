```python
import numpy as np
import matplotlib.pyplot as plt

def tri(t):
    if(abs(t) >= 0.5):
        return 0
    if(abs(t) < 0.5):
        return 1 - 2 * abs(t)

def u(t):
    if(t > 0):
        return 1
    if(t <= 0):
        return 0

def rect(t):
    if(abs(t) < 0.5):
        return 1
    if(abs(t) >= 0.5):
        return 0

def sinc(t):
    if(t == 0):
        return 1
    return np.sin(t) / t
def my(t):
    return tri(t / 2)
def my2(t):
    return my(t) + my(t - 0.5) + my(t - 1) + my(t + 0.5) + my(t + 1) 
```

```python
# xs = np.linspace(-10,10,21)
# ys = np.array()
xs = []
ys = []
for i in range(-100,120):
    xs.append(i * 0.1)
    ys.append(my2(i * 0.1))
plt.plot(xs,ys)
plt.xticks( np.linspace(-10,10,11,endpoint = True))
plt.grid(True)
plt.show()
```
